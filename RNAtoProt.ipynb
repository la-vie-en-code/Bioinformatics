```python
def RNAtoProt(x):
    #Set AminoAcid to RNAcodon correspondance from internet
    Dicodon = {"UUU" : "F", "CUU" : "L", "AUU" : "I", "GUU" : "V","UUC" : "F", "CUC" : "L", "AUC" : "I", "GUC" : "V","UUA" : "L", "CUA" : "L", "AUA" : "I", "GUA" : "V","UUG" : "L", "CUG" : "L", "AUG" : "M", "GUG" : "V","UCU" : "S", "CCU" : "P", "ACU" : "T", "GCU" : "A","UCC" : "S", "CCC" : "P", "ACC" : "T", "GCC" : "A","UCA" : "S", "CCA" : "P", "ACA" : "T", "GCA" : "A","UCG" : "S", "CCG" : "P", "ACG" : "T", "GCG" : "A","UAU" : "Y", "CAU" : "H", "AAU" : "N", "GAU" : "D","UAC" : "Y", "CAC" : "H", "AAC" : "N", "GAC" : "D","UAA" : "STOP", "CAA" : "Q", "AAA" : "K", "GAA" : "E","UAG" : "STOP", "CAG" : "Q", "AAG" : "K", "GAG" : "E","UGU" : "C", "CGU" : "R", "AGU" : "S", "GGU" : "G","UGC" : "C", "CGC" : "R", "AGC" : "S", "GGC" : "G","UGA" : "STOP", "CGA" : "R", "AGA" : "R", "GGA" : "G","UGG" : "W", "CGG" : "R", "AGG" : "R", "GGG" : "G"}
    #create list for the sequence of AA
    AA = ""
    #polymerize AAseq
    for i in range(0, len(x)-(3+len(x)%3), 3):
        if Dicodon[x[i:i+3]] == "stop":
            break
        AA += Dicodon[x[i:i+3]]
    # Print AA
    return AA
```

```python
File = open("rosalind_prot-9.txt","r")
Seq = File.read()
AAs = str(RNAtoProt(Seq))
```
